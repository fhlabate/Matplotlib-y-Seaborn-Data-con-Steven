```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Datos
x = np.linspace(0,10,100)
y = np.sin(x)

#Crear figura y eje
fig, ax = plt.subplots() #Frame donde van a entrar todos los datos
ax.plot(x,y, color = 'white')

#Personalizar el gráfico
ax.set_title('Gráfico de Seno')
ax.set_xlabel('X')
ax.set_ylabel('Seno')

#Mostrar el gráfico
plt.show()
```

## [Parte 4 | Tipos de Gráficos](https://www.youtube.com/watch?v=-gOs-NcsBzo&list=PLEiS2zOIb3riJOZzZEK_vuvO-AItS9Gy5&index=4)

### Gráfico de Barras

```python
fig, ax = plt.subplots()
categorias = ['Categoria A', 'Categoria B', 'Categoria C']
valores = [10, 15, 7]

ax.bar(categorias, valores)
ax.set_title('Grafico barras')
ax.set_xlabel('Categorias')
ax.set_ylabel('valores')
plt.show()
```

### Gráfico de dispersión

```python
x = np.random.rand(50)
y = np.random.rand(50)

fig,ax = plt.subplots()
ax.scatter(x,y)
ax.set_title('Gráfico dispersión')
ax.set_xlabel('Variable X')
ax.set_ylabel('Variable Y')
plt.show()
```

### Histograma (Gráfico de densidad)

```python
data = np.random.randn(100000)

fig, ax = plt.subplots()
ax.hist(data, bins = 1000)
ax.set_title ('Histograma')
ax.set_xlabel('Valores')
ax.set_ylabel('Frencuencia')
plt.show()
```

## [Parte 5 | Personalización de Gráficos](https://www.youtube.com/watch?v=s-ZcRTrTOvg&list=PLEiS2zOIb3riJOZzZEK_vuvO-AItS9Gy5&index=6)

```python
x = [0,1,1.5]
y = [0,1,1.5]
fig, ax = plt.subplots()
ax.plot(x,y, color = 'red', linestyle = '--', marker = 'o')
ax.set_title('Personalizado')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### Anotaciones

```python
fig, ax = plt.subplots()
ax.plot(x,y, alpha = 0.5) #alpha = transparencia
ax.text(0.5,0.8, 'Punto central', transform = ax.transAxes)
ax.annotate('Este punto es clave', xy=(0.6, 0.6), xytext = (1.2,0.7), #Marcadores
             arrowprops = dict(facecolor = 'Green', shrink = 2))
ax.set_title('Gráfico con Anotaciones')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

## [Parte 6 | Personalización avanzada](https://www.youtube.com/watch?v=_eJVsMP0LoE&list=PLEiS2zOIb3riJOZzZEK_vuvO-AItS9Gy5&index=6)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Datos
x = np.linspace(0,10,100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,y, marker = 'o', linestyle = '-', color = 'Green')

#Limitación de ejes
    #Limitación de eje x
ax.set_xlim([0,5])
    #Limitación de eje x
ax.set_ylim([-1,1])

ax.set_title('Gráfico de Seno')
ax.set_xlabel('X')
ax.set_ylabel('Seno')

plt.show()
```

### Creación de dos gráficos

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15,4))
ax1.plot(x,y)
ax1.set_title('Gráfico 1')
ax2.plot(x, np.cos(x))
ax2.set_title('Gráfico 2')
plt.show()
```

### Estilo "ggplot"

```python
plt.style.use('ggplot')
x = np.linspace(0,10,100)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Gráfico GGplot', color = 'Black')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

###  Dataset de frutos secos
[Dataset](https://drive.google.com/drive/folders/1uqy_EtGaMN2_LJHAZsu3x4c4PBYxe5ZA)

#### Crear un gráfico de barras para comparar las cantidades vendidas de diferentes tipos de frutos secos.

```python
data = pd.read_csv('Datos_Empresa_Frutos_Secos.csv', delimiter = ',')
ventas_por_producto = data.groupby('Tipo Producto')['Cantidad Vendida (kg)'].sum().sort_values(ascending = False)

plt.style.use('dark_background')
fig, ax = plt.subplots()
ventas_por_producto.plot(kind = 'bar', color = "Skyblue", ax = ax)
ax.set_title('Cantidad vendida por tipo de producto', fontsize = 14)
ax.set_xlabel('Tipo de producto', fontsize = 12)
ax.set_ylabel('Cantidad vendida (kg)', fontsize = 12)

plt.show()
```

## [Parte 8 | Segundo reto](https://www.youtube.com/watch?v=Sdzg5ceIUhs&list=PLEiS2zOIb3riJOZzZEK_vuvO-AItS9Gy5&index=9)

#### Visualizando la evolución del precio de las Almendras

```python
data = pd.read_csv('Datos_Empresa_Frutos_Secos.csv', delimiter = ',')
data['Fecha Venta'] = pd.to_datetime(data['Fecha Venta'], format = '%d/%m/%Y')

almendras = data[data['Tipo Producto'] == 'Almendras']
almendras.head()
```

```python
almendras['Precio (€/kg)'] = almendras['Precio (€/kg)'].str.replace(',', '.')
almendras['Precio (€/kg)'] = almendras['Precio (€/kg)'].astype(float)

precio_promedio_almendras = almendras.groupby('Fecha Venta')['Precio (€/kg)'].mean()

fig,ax = plt.subplots()

precio_promedio_almendras.plot(kind = 'line', color = 'lightgreen', ax = ax)

ax.set_title('Evolución Precio Almendra')
ax.set_xlabel('Fecha')
ax.set_ylabel('Precio')
plt.show()
```

#### Combinando Matplotlib con Pandas y Seaborn
- **Pandas** -> Manera muy simple de sacar gráficos por pantalla y sacar conclusiones rápidas. Entrenamiento de un algoritmo, entender ventas.
- **Seaborn** -> Permite sacar gráficos estéticos más rápido que Matplotlib. (Seaborn no tiene más estética, lo que tiene es una configuración por defecto estéticamente más agradable)

### Pandas

```python

data = pd.DataFrame({
    'A' : range(1,6),
    'B' : [x**2 for x in range(1,6)]
})
data.plot()
```

### Matplotlib

```python
fig, ax = plt.subplots()
data.plot(x='A', y ='B', ax = ax)
plt.show()
```

### Seaborn

```python
import seaborn as sns
data = np.random.randn(10000)
plt.figure(figsize = (8,6))
sns.histplot(data, kde =  True, bins = 30) #Kde = Línea de densidad (Evolución de la distribución)
plt.title('Histograma con Seaborn')
plt.xlabel('Valores')
plt.ylabel('Densidad')
```
